```python
def delete_item(self, data):
    if self.is_empty():
        return
    
    temp = self.last

    # Case: Only one element in the list
    if temp == temp.next and temp.item == data:
        self.last = None
        return

    # Traverse the list to find and delete the node
    while True:
        if temp.next.item == data:
            # Deleting the item
            if temp.next == self.last:  # If the node to be deleted is the last node
                temp.next = self.last.next
                self.last = temp
            else:
                temp.next = temp.next.next
            break  # Exit after deleting the item
        temp = temp.next

        # If we have traversed the whole list
        if temp == self.last:
            break
```
